## combine chars 

```python
from decimal import *

def combine_chars(filtered_df_f):
    #filtered_df_f = chars_df[chars_df['size'] == Decimal('8.998')]
    filtered_dict_f = filtered_df_f.to_dict(orient='records')


    words = []
    x0 =[]
    x1 =[]
    top = []
    bottom =[]
    dictlist = []

    check = 0 
    combined_words = ""
    for row in filtered_df_f.itertuples():
        
        if(abs(row.x1_minus_x0) < 1): ## if it's part of the same word, can later be assigned as a coefficient
            combined_words += row.text + ""   ##combine chars into a word
            if(check == 0):
                x0.append(row.x0)
                top.append(row.top)
                bottom.append(row.bottom)
                check = 1
        else:                        ## new word
            x1.append(row.x1)
            words.append(combined_words)
            combined_words = ""
            check = 0
            
    zipbObj = list(zip(bottom, words,top, x0,x1))
    listOfStr = ["bottom", "text" , "top" , "x0" , "x1" ]
    
    for i in zipbObj:
        zipb = zip(listOfStr, i)
        dictOfWords = dict(zipb)
        dictlist.append(dictOfWords)
    
   
    return dictlist
```

```python
fonts = chars_df['fontname'].unique()  # different font types
sizes = chars_df['size'].unique()       # different font sizes
```

```python
from decimal import *


for size in sizes:
  
    filtered_df = chars_df[chars_df['size'] == size]

    s = combine_chars(filtered_df)
   # print(s)
    im = p0.to_image(resolution=300)   
    im.draw_rects(s)
    display(im)
  
```
