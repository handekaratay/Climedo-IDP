### combine words

```python
### TODO: iki satÄ±r olanlar??

def combineWords(df):
    ### convert df to pdfplumber dict
    dict_objects = df.to_dict(orient='records')
    
    ### obtain overall bbox
    x0, top, x1, bottom = pdfplumber.utils.objects_to_bbox(dict_objects)
    
    ### combine words
    combined_txt = ""
    for text in df.text:
        combined_txt += text + " "
    combined_txt = combined_txt.strip()
        
    return {"x0": x0,"x1": x1,"top": top,"bottom": bottom,"text": combined_txt}
        
```

```python
sameline_words = []

for line in lines:#  
    
    ### select rows with those unique lines
    filter_df = words_df[words_df['top'] == line]   ### df type
    
    ### combine them
    w = combineWords(filter_df)
    sameline_words.append(w)
##split by space that is greater than bir space boyutu
```
